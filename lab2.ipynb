Заметьте, что столбец `DebtRatio` содержит неправдоподобные данные. Только значения, соответствующие известному месячному доходу, являются отношениями. Остальные – абсолютные значения месячных выплат процентов.

Исправьте данные, сделав все значения столбца `DebtRatio` абсолютными (умножьте их на `MonthlyIncome`).  Чтобы ваша программа быстро работала на полных данных, постарайтесь не использовать цикл.

#### *Функции, которые могут пригодиться при решении:*

Обращение к элементам DataFrame:
  * элемент: `data.loc[i, 'названиеСтолбца']`
  * столбец: `data['названиеСтолбца']`
  * подматрица: `data.loc[a:b, списокНазванийСтолбцов]`

Условная индексация:
* `data.loc[data['столбец'] > 20, списокНазванийСтолбцов]`

лучше писать так:

* `i = data['столбец'] > 20`  # вектор `True` и `False`
* `data.loc[i, 'названиеСтолбца']`

У подматриц номера строк наследуются от исходной.

* `pandas.isnull(скаляр или массив)` - проверка, является ли значение неопределенным (`NaN`)
* `pandas.notnull(скаляр или массив)` - проверка, является ли значение определенным (не `NaN`)

````python
# Создаем условие для определения, какие значения DebtRatio нужно исправить
i = pd.notnull(data['MonthlyIncome'])

# Умножаем значения столбца DebtRatio на значения столбца MonthlyIncome там, где i == True
data.loc[i, 'DebtRatio'] = data['DebtRatio'] * data['MonthlyIncome']
````

### 9a

Постройте график рассеяния на осях `age` и `Debt`. Cиним отметьте клиентов без серьезных задолженностей (`SeriousDlqin2yrs = 0`) и красным — должников (`SeriousDlqin2yrs = 1`).

````python
# Разделяем данные по задолженностям
no_debt = data[data['SeriousDlqin2yrs'] == 0]
debt = data[data['SeriousDlqin2yrs'] == 1]

# Строим график рассеяния
plt.scatter(no_debt['age'], no_debt['AbsoluteDebt'], color='blue', label='No Debt')
plt.scatter(debt['age'], debt['AbsoluteDebt'], color='red', label='Debt')

plt.xlabel('Age')
plt.ylabel('Debt')
plt.legend()
plt.show()
````

### 9b
Постройте на одном графике две **нормированные** плотности распределения: красную – для месячного дохода клиентов с задолженностями, синюю – для месячного дохода клиентов без задолженностей. По оси абсцисс отобразите значения до 25000.

````python
# Фильтрация данных по месячному доходу и задолженностям
income_debt = data[data['MonthlyIncome'] <= 25000]

# Разделение данных на две группы: с задолженностями и без задолженностей
income_with_debt = income_debt[income_debt['SeriousDlqin2yrs'] == 1]['MonthlyIncome']
income_without_debt = income_debt[income_debt['SeriousDlqin2yrs'] == 0]['MonthlyIncome']

# Строим нормированные плотности распределения
sns.kdeplot(income_without_debt, color='blue', label='No Debt')
sns.kdeplot(income_with_debt, color='red', label='Debt')

plt.xlim(0, 25000)
plt.xlabel('Monthly Income')
plt.ylabel('Density')
plt.legend()
plt.show()
````

### 9c*

Какие закономерности вы можете наблюдать на получившихся графиках?

````python
# Отфильтруем данные по месячному доходу до 25000
filtered_data = data[data['MonthlyIncome'] <= 25000]

# Выберем только нужные столбцы для анализа
selected_columns = ['age', 'MonthlyIncome', 'NumberOfDependents']

# Построим попарные диаграммы рассеяния
pd.plotting.scatter_matrix(filtered_data[selected_columns], diagonal='hist', figsize=(15, 10))
plt.show()
````
